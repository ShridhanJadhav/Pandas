## Map

The map method lets us map values in a column to other values. Let's use map to create a binary column that is 1 for females and 0 for males. We give the mapping that we want by providing the appropriate dictionary

```python
df_titanic.head()
```

```python
len(df_titanic)
```

```python
for i in range(len(df_titanic)) :
    s = df_titanic.loc[i,"Gender"]
    print(s)
    if s == True:
        df_titanic.loc[i,"Gender"] = "Male"
    else:
        df_titanic.loc[i,"Gender"] = "Female"
df_titanic.head()
```

```python
def newgender(x):
    if x == "male":
        return "0"
    else:
        return "1"
```

```python
newgender("male")
```

```python
for i in range(len(df_titanic)):
    s = df_titanic.loc[i,"Gender"]
    df_titanic.loc[i,"Gender_Function"] = newgender(s)

df_titanic.head()
```

```python
df_titanic["Gender_Apply"] = df_titanic["Sex"].apply(newgender)
df_titanic.head()
```

```python
df_titanic["Gender_lambda"] = df_titanic["Sex"].apply(lambda x: 0 if x=="male" else 1)
df_titanic.head()
```
